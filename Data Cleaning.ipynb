```python
import numpy as np
```

```python
data = np.array([['Alice', 25, 'New York'], ['?', 37, 'Japan'], ['Lucky', 29, 'India'], ['Bob', '?', 'Los Angeles'], ['Colins', 27, 'Norland'], ['Charlie', 35, 'Chicago'],['Faith', 30, '?'],['Ducky', '?', 'Florida'],['Glory', 32, 'Spain']])
columns = ['Name', 'Age', 'City']
sf = pd.DataFrame(data, columns=columns)
print(sf)
```

```python
sf = sf.replace('?', np.nan)
sf
```

```python
sf['Name'] = sf['Name'].fillna('Sugar')
```

```python
sf
```

```python
sf['City'] = sf['City'].fillna('Qatar')
sf
```

```python
sf['Age'] = sf['Age'].fillna(method="backfill")
sf
```
